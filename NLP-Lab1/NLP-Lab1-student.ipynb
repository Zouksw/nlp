## 什么是正则表达式 (Regular Expressions)？
* 正则表达式使用某种预定义的模式去匹配一类具有共同特征的字符串，主要用于处理字符串，可以快速、准确地完成复杂的查找、替换等处理要求，在文本编辑与处理、网页爬虫之类的场合中有重要应用。
* 正则表达式就是对字符串（包括普通字符，例如，a 到 z 之间的字母）和特殊字符（称为“元字符”）操作的一种逻辑公式。
* 使用单个字符串来描述匹配一系列符合某个句法规则的字符串。

## 正则表达式搜索过程
* 正则表达式过程：依次拿出表达式和文本中的字符比较，如果每一个字符都能匹配，则匹配成功，否则匹配失败。如果有多个匹配项，则可以将搜索设计为返回行上的每个匹配项，或者只返回第一个匹配项。

写出在字符串"Woodchucks:interesting links to woodchucks and lemurs"中匹配出"woodchucks"，返回Match对象和span。

## 最简单的正则表达式是简单字符序列
![image.png](pic/2.png)

写出在字符串"Woodchucks or woodchucks"中匹配出"Woodchunks或woodchunks"，返回Match对象和span。

```python
import re

r = '[Ww]oodchucks'
match = re.search(r, 'Woodchucks or woodchucks')
print(match)
print(match.span())
```

```python
import re

r = '[Ww]oodchucks'
match = re.findall(r, 'Woodchucks or woodchucks')
print(match)
```

## 正则表达式语法——More Disjunction
###  ?  *  +  .
![image.png](pic/4.png)

在给定的文本字符串中"color,is coulour,oh,ooh,ooooooh!"搜索并提取符合正则表达式模式'oo*h'的部分。

```python
import re

r = 'oo*h'
match = re.search(r, 'color,is coulour,oh,ooh,ooooooh!')
print(match)
print(match.span())
```

## 在python中如何使用正则表达式

匹配字符串"hello Rose"以 "hi"、"hello" 或 "hey" 开头，后跟零个或多个空格，然后是零个或多个小写字母的情况,返回match对象和span。

```python
import re

text = "hello Rose"
pattern = r'(hi|hello|hey)\s*[a-z]*'

match = re.search(pattern, text)
print(match)
print(match.span())
```

将"alpha.beta....gama delta"通过一个或多个 "." 号分割，返回分割的字符。

```python
import re

text = "alpha.beta...gama delta"
pattern = r'\.+'

match = re.split(pattern, text)
print(match)
```

匹配字符串中"alpha.beta....gama delta"所有的连续的字母序列，返回连续的字母序列。

```python
import re

text = "alpha.beta....gama delta"
pattern = r'[a-zA-Z]+'

result = re.findall(pattern, text)
print(result)  
```

将文本"Dear {name}："中的 {name} 替换为 "Mr.John"。

```python
import re

text = "Dear {name}: "
pattern = r'\{name\}'
replace = 'Mr.John'

result = re.sub(pattern, replace, text)
print(result)  
```

将字符串"aaa abc abde"中的每个字符"a"替换为对应的大写字母"A"。

```python
import re

text = "aaa abc abde"
pattern = r'a'
replace = 'A'

result = re.sub(pattern, replace, text)
print(result)  
```

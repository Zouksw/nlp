1.任务描述：合法的IPv4地址由4组数字组成，每组数字之间以.分隔，且取值范围是0-255。请在下列数字中判断出合法的地址。  
输入：66666666666  
23333333  
500.1.2.5  
0.0.0.0  
192.0.0.1  
233.233.233.233  
102.52.96.3  

```python
import re
pattern = r'^((25[0-5]|2[0-4]\d|1\d{2}|[1-9]\d|[1-9])\.)((25[0-5]|2[0-4]\d|[01]?\d\d?)\.){2}(25[0-5]|2[0-4]\d|[01]?\d\d?)$'
list = ['66666666666',  
        '23333333',  
        '500.1.2.5',  
        '0.0.0.0', 
        '255.255.255.255',
        '192.0.0.1',  
        '233.233.233.233',  
        '102.52.96.3']
for item in list:
    if re.search(pattern, item) is not None:
        print(item)
```

2.任务描述：将句子中的中文日期替换为阿拉伯数字，并改为YYYY-MM-DD格式输出。  
输入：新中国于一九四九年十月一日成立。  
我的出生日期是二零零零年九月二十三日。  

```python
import re

def chinese_to_arabic(chinese_num):
    num_map = {
        '零': 0, '一': 1, '二': 2, '三': 3, '四': 4,
        '五': 5, '六': 6, '七': 7, '八': 8, '九': 9, '十': 10
    }
    
    if chinese_num == '十':
        return 10
    
    if chinese_num.startswith('十'):
        return 10 + num_map.get(chinese_num[1], 0)

    if chinese_num.endswith('十') and len(chinese_num) == 2:
        return num_map[chinese_num[0]] * 10

    if '十' in chinese_num:
        parts = chinese_num.split('十')
        if len(parts) == 2:
            tens = num_map[parts[0]] * 10 if parts[0] else 10
            ones = num_map[parts[1]] if parts[1] else 0
            return tens + ones

    result = 0
    for char in chinese_num:
        if char in num_map:
            result = result * 10 + num_map[char] if result > 0 else num_map[char]
    return result

text_list = [
    '新中国于一九四九年十月一日成立。',
    '我的出生日期是二零零零年九月二十三日。'
]

pattern = r'([一二三四五六七八九零]{2,4})年([一二三四五六七八九零十]{1,3})月([一二三四五六七八九零十]{1,3})日'

for text in text_list:
    def replace_date(match):
        year_chinese = match.group(1)
        month_chinese = match.group(2)
        day_chinese = match.group(3)

        year = ''.join([str('零一二三四五六七八九'.index(c)) for c in year_chinese])
        
        month = str(chinese_to_arabic(month_chinese))
        
        day = str(chinese_to_arabic(day_chinese))

        return f'{year}-{month}-{day}'

    result = re.sub(pattern, replace_date, text)
    print(result)
```

3.任务描述：将句子中的电话号码提取出来，电话号码格式为区号（可选）后接“-”或空格，再加十位数字。  
输入：我的电话号码是1234567890，另一个电话号码是+86-4321567890。如果以上都不可用，您可以在16点前致电+91 5645456545与我联系。

```python
import re

text = "我的电话号码是1234567890，另一个电话号码是+86-4321567890。如果以上都不可用，您可以在16点前致电+91 5645456545与我联系。"

pattern = r'(?:\+\d{1,3}[- ])?\d{10}'

phone_numbers = re.findall(pattern, text)

print("提取的电话号码：")
for phone in phone_numbers:
    print(phone)
```
